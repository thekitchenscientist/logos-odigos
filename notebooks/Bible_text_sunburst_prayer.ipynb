```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import webbrowser
import numpy as np
import plotly.io as pio
pio.renderers.default = "iframe"
```

```python
df = pd.read_csv('D:/LO_resources/eng-BSB.csv', encoding='utf-8', index_col=0)
reference_df = pd.read_csv('D:/LO_resources/bible_with_reference_matches.csv', encoding='utf-8', index_col=0)
mention_df = pd.read_csv('D:/LO_resources/bible_with_mention_matches.csv', encoding='utf-8', index_col=0)
answer_df = pd.read_csv('D:/LO_resources/bible_with_answer_matches.csv', encoding='utf-8', index_col=0)
```

```python
verse_df = df
verse_df = verse_df.astype({'verse_number':str,'chapter_number':str})
verse_df['id'] = verse_df['book'] + " - Chapter " + verse_df['chapter_number'] + " - Verse " + verse_df['verse_number']
verse_df['label'] = "Verse " + verse_df['verse_number']
verse_df['parent'] = verse_df['book'] + " - Chapter " + verse_df['chapter_number']
verse_df['value'] = verse_df['verse_word_count']
verse_df
```

```python
#need ids, labels, parents, value
#for chapter id "book - chapter number"
chapter_df = df.groupby(['alternative_book_order','alternative_group','book', 'chapter_number'], as_index=False).agg(chapter_word_count=('verse_word_count', 'sum'))
chapter_df = chapter_df.astype({'chapter_number':str})
chapter_df['id'] = chapter_df['book'] + " - Chapter " + chapter_df['chapter_number']
chapter_df['label'] = "Chapter " + chapter_df['chapter_number']
chapter_df['parent'] = chapter_df['book']
chapter_df['value'] = chapter_df['chapter_word_count']
chapter_df
```

```python
book_df = chapter_df.groupby(['alternative_book_order','alternative_group','book'], as_index=False).agg(book_word_count=('chapter_word_count', 'sum'))
book_df['id'] = book_df['book']
book_df['label'] = book_df['book']
book_df['parent'] = book_df['alternative_group']
book_df['value'] = book_df['book_word_count']
book_df
```

```python
group_df = book_df.groupby(['alternative_group'], as_index=False).agg(group_word_count=('book_word_count', 'sum'))
group_df['id'] = group_df['alternative_group']
group_df['label'] = group_df['alternative_group']
#group_df['parent'] = np.where(group_df['alternative_group']=='New Testament', 'The New Testament', 'The Old Testament')
group_df['parent'] = ''
group_df = group_df.rename(columns={'group_word_count': 'value'})
group_order = ['Covenant History','The Prophets','The Writings','New Testament']
group_df['group_order'] = group_df['alternative_group'].apply(lambda x: group_order.index(x) + 1)
group_df = group_df.sort_values(by=['group_order'], ascending=True)
group_df
```

```python
OT_count = group_df[group_df['alternative_group']!='New Testament']['value'].sum()
NT_count = group_df[group_df['alternative_group']=='New Testament']['value'].sum()
```

```python
bible_df = pd.DataFrame({'id':["The Old Testament","The New Testment"],'label':["The Old Testament","The New Testment"],'parent':["",""],'value':[OT_count,NT_count]})
bible_df
```

```python
#join them all together OT only
combined_df = pd.concat([bible_df,group_df[['id','label','parent','value']],book_df[['id','label','parent','value']],chapter_df[['id','label','parent','value']]])
combined_df.reset_index(inplace=True, drop=True)
combined_df.drop([1,5],inplace=True)
combined_df.reset_index(inplace=True, drop=True)
combined_df.drop(combined_df.index[43:70],inplace=True)
combined_df.drop(combined_df.index[972:],inplace=True)
combined_df.reset_index(inplace=True, drop=True)
OT_combined_df = combined_df
combined_df
```

```python
#join them all together NT only
combined_df = pd.concat([bible_df,group_df[['id','label','parent','value']],book_df[['id','label','parent','value']],chapter_df[['id','label','parent','value']]])
combined_df.reset_index(inplace=True, drop=True)
combined_df.drop([0,2,3,4],inplace=True)
combined_df.reset_index(inplace=True, drop=True)
combined_df.drop(combined_df.index[2:41],inplace=True)
combined_df.drop(combined_df.index[29:957],inplace=True)
combined_df.reset_index(inplace=True, drop=True)
NT_combined_df = combined_df
combined_df
```

```python
reference_color_df = reference_df.groupby(['alternative_book_order','alternative_group','book', 'chapter_number'], as_index=False).agg({'match_verse': 'sum'})
reference_color_df = reference_color_df.astype({'chapter_number':str})
reference_color_df['id'] = reference_color_df['book'] + " - Chapter " + reference_color_df['chapter_number']

mention_color_df = mention_df.groupby(['alternative_book_order','alternative_group','book', 'chapter_number'], as_index=False).agg({'match_verse': 'sum'})
mention_color_df = mention_color_df.astype({'chapter_number':str})
mention_color_df['id'] = mention_color_df['book'] + " - Chapter " + mention_color_df['chapter_number']

answer_color_df = answer_df.groupby(['alternative_book_order','alternative_group','book', 'chapter_number'], as_index=False).agg({'match_verse': 'sum'})
answer_color_df = answer_color_df.astype({'chapter_number':str})
answer_color_df['id'] = answer_color_df['book'] + " - Chapter " + answer_color_df['chapter_number']
```

```python
custom_df = verse_df
custom_df = custom_df.astype({'alternative_book_order':str,'book':str,'chapter_number':str,'text':str})
custom_df = custom_df.groupby(['alternative_book_order','book','chapter_number'], as_index=False)['text'].apply(lambda x: '<br>'.join(x))
custom_df = custom_df.astype({'alternative_book_order':int,'chapter_number':int})
custom_df = custom_df.sort_values(by=['alternative_book_order','chapter_number'], ascending=True)
custom_text_list_part_2 = custom_df['text'].tolist()
custom_data = custom_text_list_part_1+custom_text_list_part_2
```

```python
fig =go.Figure(go.Sunburst(
    ids= id_list,
    labels=label_list,
    parents=parent_list,
    values=value_list,
    marker=dict(colors=colors),
    customdata=custom_data,
    branchvalues="total",
    name='',
))
#fig.update_layout(template="seaborn") #["plotly", "plotly_white", "plotly_dark", "ggplot2", "seaborn", "simple_white", "none"]
fig.update_layout(margin = dict(t=0, l=0, r=0, b=0))
fig.update_traces(sort=False, selector=dict(type='sunburst'))
fig.update_traces(
    hovertemplate=(
        '%{customdata}'
    )
)
fig.update_layout(
    title_text='The Bible in a Sunburst',
    title_font=dict(size=24),
    title_x=0.025,
    title_y=0.95,
    title_xanchor='left',
    title_yanchor='top',
)
fig.add_annotation(
            text="""This graphic contains the entire bible. Hover over a Chapter to read it.<br>
            Click on a section to focus on that book or group of books.<br>
            The sections are sized by the total number of words.""",
            font=dict(size=14),
            x=0.025,
            y=0.925,
            xref='paper',
            yref='paper',
            showarrow=False,
            align='left',
        )
fig.add_annotation(
            text="""The Holy Bible, Berean Standard Bible, BSB is produced in cooperation with Bible Hub,<br> 
            Discovery Bible, OpenBible.com, and the Berean Bible Translation Committee.<br>
            This text of God's Word has been dedicated to the public domain (April 2023).""",
            font=dict(size=12),
            x=0.98,
            y=0.025,
            xref='paper',
            yref='paper',
            showarrow=False,
            align='left',
)
fig.show()
fig.write_html("D:/LO_resources/sunburst_navigation_coloured.html")
```
